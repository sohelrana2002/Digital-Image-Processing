```python
import PIL.Image as image # open image
import matplotlib.pyplot as plt # for plot to show image
```

a)	Read an image.

```python
img = image.open('Images/download.png')
```

b)	Display the image resolution in the format: Resolution = Width × Height pixels.

```python
width, height = img.size
print(f"Resolution = {width} × {height} pixels")
```

Print the image color mode.

```python
print(f"Color mode = {img.mode}")
```

Convert the image into grayscale

```python
img_gray = img.convert('L')
```

Display both the original image and the grayscale image

```python
plt.figure(figsize=(15, 10))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

Label each displayed image with an appropriate title & Save the grayscale image

```python
gray_file = 'Images/download_gray.png'
img_gray.save(gray_file)
```

```python
import numpy as np
import matplotlib .pyplot as plt
```

Read a grayscale image.

```python
img = image.open('Images/download_gray.png')
```

Increase the brightness of the image by multiplying each pixel value by a scaling factor S₁ = 1.5

```python
gray_array = np.array(img)
S1 = 1.5
bright_high = np.clip(gray_array * S1, 0, 
255).astype("uint8")
```

Decrease the brightness of the image by multiplying each pixel value by a scaling factor S₂ = 0.3

```python
S2 = 0.3
bright_low = np.clip(gray_array * S2, 0, 
255).astype("uint8")
```

Compare the original grayscale image with the brightness-adjusted images.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(gray_array, cmap="gray")
plt.title("Original grayscale")

plt.subplot(1, 3, 2)
plt.imshow(bright_high, cmap="gray")
plt.title("Brightness increased (S = 1.5)")

plt.subplot(1, 3, 3)
plt.imshow(bright_low, cmap="gray")
plt.title("Brightness decreased (S = 0.3)")
plt.show()
```
