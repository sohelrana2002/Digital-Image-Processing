```python
import PIL.Image as image # open image
import matplotlib.pyplot as plt # for plot to show image
import numpy as np # make array and perform array operation
```

a) Read a grayscale image

```python
img = image.open("Images/hand_xray.jpg")
plt.imshow(img, cmap='gray')
plt.title("Original Gray scale Image")
plt.axis("off")
plt.show()
```

Set Height and Width

```python
img_array = np.array(img)

#image size
H, W = img_array.shape
print(f"Resolution = {W} x {H} pixels")
```

b) Set a threshold value, k = 128

c) Apply binary thresholding using


```python
k = 128

binary_img = np.where(img_array > k, 255, 0)
```

d)	Display the binary image

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Gray scale Image")

plt.subplot(1, 2, 2)
plt.imshow(binary_img , cmap='gray')
plt.title("Binary Image (k=128)")
plt.axis("off")
plt.show()
```

e)	Create three binary images using different thresholds: k = 50, 100, 150

f)	Compare the results

```python
thresholds = [50, 100, 150]

plt.figure(figsize=(12,4))

for i, t in enumerate(thresholds):
    binary_images = np.where(img_array > t, 255, 0)

    plt.subplot(1, 3, i+1)
    plt.title(f"k = {t}")
    plt.imshow(binary_images, cmap="gray")
    plt.axis("off")

plt.show()
```
