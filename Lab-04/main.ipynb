```python
import PIL.Image as image # open image
import matplotlib.pyplot as plt # for plot to show image
import numpy as np # make array and perform array operation
```

a)	Read a grayscale image

```python
img = image.open("Images/hand_xray.jpg").convert("L")
gray_array = np.array(img)
plt.imshow(gray_array, cmap='gray')
plt.title("Original Gray scale Image")
plt.show()
```

b)	Apply image negative transformation using the formula:  s = 255 – r

```python
negative_array = 255 - gray_array
```

c)	Display both images

```python
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(gray_array, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

# Negative Image
plt.subplot(1, 2, 2)
plt.imshow(negative_array, cmap='gray')
plt.title("Negative Image (s = 255 - r)")
plt.axis("off")

plt.show()
```
